# Notebook Description

This notebook aims at creating a subset of the HP fanfiction dataset, focused on English stories in order to run analysis on the  summary of the story.
This will also create a smaller csv file which will be easier to upload & download.

```python
# Import libraries

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # data visualisation
import seaborn as sns # data visualisation plus
```

# Import the file & visualise the data
We will import the csv file found on Kaggle and visualise the data.
Later on, we can create a Database with SQLLite or MySQL to query the data more easily.

```python
df = pd.read_csv("hpcleanvlarge1.csv", 
                 thousands=',') # this parameters setup the expected values for identifying large numbers (in american format for numbers))

df.head()
```

# Identify stories in English
All stories are tagged by their languages. We want to focus only on English stories for the moment to run a first language analysis on the summary.

```python
eng_df = df.query("language == 'English'").copy() # create a copy of the main dataframe with only stories in English language.

eng_df.head()
```

```python
# verification if the query worked as expected
eng_df['language'].unique()
```

# Save to csv
Export the file in order to use it in various context to run different analysis. We will use the zipped format which will simplify the uploading on Git (Pandas has a build-in compression and decompression feature).

```python
eng_df.to_csv('hpclean_engl.csv.gz')
```
